```python
# import des librairies
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import des données
df = pd.read_csv('train.csv', sep=',')
df.head()
```

```python
#phase d'exploration
df.info()
```

```python
# definir x et y en commencant par la deuxieme colonne
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
#phase d'exploration
df.describe()
```

```python
# selectionner la premiere ligne du fichier train csv 
img = df.iloc[3]

img

# enlever la premier colonne de img
img = img.drop('label')
img

# convertir img en array
img = np.array(img)
img
```

```python
# convertir img en image
img = img.reshape(28,28)
img
```

```python
# afficher l'image en niveau de gris
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# afficher les images de 1 a 9 dans une seule figure 
for i in range(9):
    img = df.iloc[i]
    img = img.drop('label')
    img = np.array(img)
    img = img.reshape(28,28)
    plt.subplot(330+1+i)
    plt.imshow(img)
plt.show()
    
```

```python
# splite data 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print (x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# implementer le model SVM 
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(x_train, y_train)
```

```python
# autre facon de faire
#model = SVC(kernel='poly', degree=3)
#model.fit(x_train, y_train)
```

```python
# autre facon de faire
#model = SVC(kernel='rbf', gamma=0.1, C=100)
#model.fit(x_train, y_train)
```

```python
# prediction
pred = model.predict(x_test)

#afficher l'accuracy_score
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, pred)
print(accuracy)

#precision score
from sklearn.metrics import precision_score
precision = precision_score(y_test, pred, average='macro')
print(precision)

#recall score
from sklearn.metrics import recall_score
recall = recall_score(y_test, pred, average='macro')
print(recall)

# classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

```python
# evaluation du model
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred)
cm
```
